```python
import pandas as pd
import matplotlib.pyplot as plt

# Your dataset
df = pd.read_csv("companies_data.csv")

# Target companies in your chosen sequence
target_companies = [
    "TCS", "Accenture", "ICICI Bank", "Cognizant", "HDFC Bank",
    "Infosys", "L&T", "Capgemine", "HCL Technologies"
]

# Filter and preserve order
df_filtered = df[df['name'].isin(target_companies)].copy()
df_filtered['name'] = pd.Categorical(df_filtered['name'], categories=target_companies, ordered=True)

# Group by company and calculate average rating
ratings = df_filtered.groupby('name')['rating'].mean().dropna()

# Prepare lists
company_names = ratings.index.tolist()
average_ratings = ratings.values.tolist()

# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(company_names, average_ratings, marker='o', linestyle='-', color='teal')
plt.title('Average Ratings of Selected Companies (Line Plot)')
plt.xlabel('Company Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Vertical Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(company_names, average_ratings, color='skyblue')
plt.title('Average Ratings of Selected Companies (Vertical Bar Plot)')
plt.xlabel('Company Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Horizontal Bar Plot
plt.figure(figsize=(10, 6))
plt.barh(company_names, average_ratings, color='orange')
plt.title('Average Ratings of Selected Companies (Horizontal Bar Plot)')
plt.xlabel('Average Rating')
plt.ylabel('Company Name')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Drop rows with missing values in the rating column
df.dropna(subset=['rating'], inplace=True)

# Define dynamic bin edges based on min and max ratings
min_rating = df['rating'].min()
max_rating = df['rating'].max()
bins = np.linspace(min_rating, max_rating, 11)  # 10 equal-width bins

# Plot a clean histogram of the 'rating' column
plt.figure(figsize=(10, 6))
plt.hist(df['rating'], bins=bins, color='blue', edgecolor='black', label='Rating')

plt.title('Distribution of Company Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data matching your sketch
data = {
    'name': ['TCS', 'TCS', 'Accenture', 'Accenture', 'Accenture', 'Infosys', 'Infosys', 'Infosys', 'HCL', 'HCL', 'HCL'],
    'employee': [10000, 12000, 20000, 30000, 50000, 25000, 40000, 60000, 8000, 15000, 22000],
    'rating': [5.0, 5.0, 4.5, 4.5, 4.5, 4.0, 4.0, 4.0, 3.5, 3.5, 3.5]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))

# Plot each company with a different color
for company in df['name'].unique():
    subset = df[df['name'] == company]
    plt.scatter(subset['employee'], subset['rating'], label=company, alpha=0.8, edgecolor='black')

# Formatting
plt.title('Scatter Plot: Employees vs Rating by Company')
plt.xlabel('Number of Employees')
plt.ylabel('Company Rating')
plt.legend(title='Company')
plt.grid(True)
plt.tight_layout()
plt.show()



```
